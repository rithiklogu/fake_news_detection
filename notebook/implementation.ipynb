```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer, StandardScaler, LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from scipy.sparse import hstack


df.drop("id", axis=1, inplace=True)

label_order = {'pants-fire': 0, 'false': 1, 'barely-true': 2, 'half-true': 3, 'mostly-true': 4, 'true': 5}
df['label'] = df['label'].map(label_order)

statement_tfidf = TfidfVectorizer(max_features=70)
X_statement = statement_tfidf.fit_transform(df['statement'])

df['subject'] = df['subject'].apply(lambda x: x.split(','))
all_subjects = pd.Series([s for sublist in df['subject'] for s in sublist])
top_subjects = all_subjects.value_counts().nlargest(10).index
df['subject'] = df['subject'].apply(lambda lst: [s for s in lst if s in top_subjects])
mlb = MultiLabelBinarizer(classes=top_subjects)
X_subject = mlb.fit_transform(df['subject'])

for col in ['speaker', 'state_info']:
    freq_map = df[col].value_counts().to_dict()
    df[col + '_freq'] = df[col].map(freq_map)

job_tfidf = TfidfVectorizer(max_features=10)
X_job = job_tfidf.fit_transform(df['job_title'])

le = LabelEncoder()
df["party_affiliation"] = le.fit_transform(df["party_affiliation"].astype(str))

count_cols = ['barely_true_counts', 'false_counts', 'half_true_counts', 'mostly_true_counts', 'pants_on_fire_counts']
scaler = StandardScaler()
df[count_cols] = scaler.fit_transform(df[count_cols])

context_tfidf = TfidfVectorizer(max_features=10)
X_context = context_tfidf.fit_transform(df['context'])

X_final = hstack([X_statement, X_subject, X_job, X_context])
X_final_dense = pd.DataFrame(X_final.toarray())

final_df = pd.concat([X_final_dense,
                      df[['speaker_freq', 'state_info_freq']],
                      df[count_cols],
                      df[["party_affiliation"]]
                     ], axis=1)

y = df['label']

```

```python
#final model
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
from sklearn.utils.class_weight import compute_sample_weight
from xgboost import XGBClassifier

# Load data
df = pd.read_csv(r"C:\Users\rithi\Desktop\GEN_AI\fake_news_detection\src\liar_data\csv file\train.csv")

# Fill missing categorical values with mode
categorical_cols = ["speaker", "job_title", "state_info", "party_affiliation", "context"]
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Drop ID column
if "id" in df.columns:
    df.drop(columns=["id"], inplace=True)

# Normalize label values
df["label"] = df["label"].astype(str).str.lower().str.strip()

# Convert labels to binary
def simplify_label(label):
    label_map = {
        "pants-fire": 0,
        "false": 0,
        "barely-true": 0,
        "half-true": 0,
        "mostly-true": 1,
        "true": 1
    }
    return label_map.get(label, -1)

df["label"] = df["label"].apply(simplify_label)
df = df[df["label"] != -1]

# Combine text fields
df["text"] = df["statement"].fillna('') + " " + df["context"].fillna('') + " " + df["job_title"].fillna('')

# TF-IDF vectorization
tfidf = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf.fit_transform(df["text"])

# Dimensionality reduction
svd = TruncatedSVD(n_components=100, random_state=42)
X_svd = svd.fit_transform(X_tfidf)

# Encode categorical features
encoded_cols = []
for col in ['speaker', 'state_info', 'party_affiliation']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoded_cols.append(col)

# Numeric features
count_cols = ['barely_true_counts', 'false_counts', 'half_true_counts',
              'mostly_true_counts', 'pants_on_fire_counts']
X_numeric = df[count_cols + encoded_cols].astype(float).values

# Final feature matrix and labels
X_final = np.hstack([X_svd, X_numeric])
y = df["label"]

# Train/Validation/Test split
X_temp, X_test, y_temp, y_test = train_test_split(X_final, y, test_size=0.2, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=42)

# Handle imbalance
sample_weights = compute_sample_weight(class_weight="balanced", y=y_train)

# Train XGBoost
model = XGBClassifier(
    objective="binary:logistic",
    eval_metric="logloss",
    learning_rate=0.1,
    max_depth=6,
    n_estimators=200,
    subsample=0.8,
    colsample_bytree=0.8,
    use_label_encoder=False,
    random_state=42
)
model.fit(X_train, y_train, sample_weight=sample_weights)

# Define evaluation
def evaluate(name, y_true, y_pred):
    print(f"\n🔹 {name} Set")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))
    print("Classification Report:\n", classification_report(y_true, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))

# Predict without threshold
y_pred_train = model.predict(X_train)
y_pred_val = model.predict(X_val)
y_pred_test = model.predict(X_test)

# Evaluate normal predictions
evaluate("Train", y_train, y_pred_train)
evaluate("Validation", y_val, y_pred_val)
evaluate("Test", y_test, y_pred_test)

# 🔻 Threshold tuning
y_probs_test = model.predict_proba(X_test)[:, 1]  # Probabilities for class 1
y_pred_thresh_test = (y_probs_test >= 0.4).astype(int)  # Adjust threshold here (e.g., 0.4)

# Evaluate with threshold tuning
evaluate("Test (Threshold=0.4)", y_test, y_pred_thresh_test)
```

```python
from sklearn.metrics import f1_score

best_thresh = 0.5
best_f1 = 0

for t in np.arange(0.1, 0.9, 0.05):
    preds = (y_probs_test >= t).astype(int)
    score = f1_score(y_test, preds)
    if score > best_f1:
        best_f1 = score
        best_thresh = t

print("Best Threshold:", best_thresh)
print("Best F1 Score:", best_f1)
```

```python
# Use best threshold
final_thresh = 0.35
y_pred_thresh_test = (y_probs_test >= final_thresh).astype(int)

evaluate("Test (Threshold=0.35)", y_test, y_pred_thresh_test)
```
